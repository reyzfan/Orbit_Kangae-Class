```python
#import library
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
%matplotlib inline
```

```python
#download a file dataset iris.csv from github
!wget -O iris.csv https://raw.githubusercontent.com/achmatim/data-mining/main/Dataset/iris.csv 
```

```python
#df.head function is read and show file in iris.csv
df = pd.read_csv('iris.csv') 
#show the five first column from dataframe
df.head() 
```

```python
#df['Label] and value_counts function show data per label and sum of value in iris.csv
df['Label'].value_counts() 
```

```python
#df.hist function show the data histogram in column SepalLengthCm with range of value is 50
df.hist(column='SepalLengthCm', bins=50) 
```

```python
#df.column function is format the dataset to group attribute using x variable
df.columns 
```

```python
#save dataframe with column on variable name x 
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] .values 
#x[0:5] function show value of number with five column in values with range 0 to 4
x[0:5] 
```

```python
#save dataframe with column name of Label as y variable
y = df['Label'].values 
#show the array y variable with index of range 0 to 4
y[0:5] 
```

```python
#save standardization result using method of StandarScaler()
x = preprocessing.StandardScaler().fit(x).transform(x.astype(float))

x[0:5]
```

```python
from sklearn.model_selection import train_test_split #import train_test_split from sklearn.model_selection
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=4) #x,y train & test function it is Divides an array or matrix into random trains and test subsets with random state 
#show the values of trains set and tests set
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
#import KNN Classifier from sklearn
from sklearn.neighbors import KNeighborsClassifier 
```

```python
k = 4
#Train Model and Predict  
knn = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train) #describes of KNeighborsClassifier is knn value
knn
```

```python
#yhat representation of the value of knn.predict in x train
yhat = knn.predict(x_test) 
#yhat is an array to show the values in yhat from index 0 to 4
yhat[0:5] 
```

```python
#import metrics from sklearn
from sklearn import metrics 
#show the metrics accuracy score from y train and value of knn.predict is x train
print("Train set Accuracy: ", metrics.accuracy_score(y_train, knn.predict(x_train))) 
#show the metrics accuracy score from y test and value of yhat from before cell
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat)) 
```

```python
Ks = 10 #save value of 10 with variable name is "ks"
mean_acc = np.zeros((Ks-1)) 
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(x_train,y_train)
    yhat=neigh.predict(x_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc #show the values of variable mean_acc with array accuracy
```

```python
plt.plot(range(1,Ks),mean_acc,'g') #show the visualization accuracy graph
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show() #show the graph result 
```

```python
#show best accuracy with max mean value and maximalized k mean accuracy
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```
